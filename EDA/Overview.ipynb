```python
def overview(dataset):
    print('Shape of dataset:')
    print(dataset.shape)
    print('Columns:')
    print(dataset.columns)
    print('First two lines:')
    print(dataset.head(2))
    print('Info:')
    print(dataset.info())
    print('Missing values:')
    print(dataset.isna().sum())
    print('Numbers of unique features:')
    print(dataset.nunique())
    return
```
